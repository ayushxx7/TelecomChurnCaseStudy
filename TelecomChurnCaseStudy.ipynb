### Filtering High Value Customers

##### Checking all columns which contain the string "rech" (short for recharge)

```python
df[df['total_rech_data_6'] > 1][['total_rech_num_6', 'total_rech_amt_6', 'max_rech_amt_6', 'date_of_last_rech_6', 'total_rech_data_6', 'max_rech_data_6', 'av_rech_amt_data_6']]
```

##### After carefully analyzing the data, it was found that the missing values for the columns represented that no recharge was done during that period. Hence, we will be filling those values with 0.

```python
df[['av_rech_amt_data_6', 'total_rech_data_6',
   'av_rech_amt_data_7', 'total_rech_data_7',
   'av_rech_amt_data_8', 'total_rech_data_8',
   'av_rech_amt_data_9', 'total_rech_data_9',
   'total_rech_amt_6', 'total_rech_amt_7',
    'total_rech_amt_8', 'total_rech_amt_9'
   ]] = df[['av_rech_amt_data_6', 'total_rech_data_6',
   'av_rech_amt_data_7', 'total_rech_data_7',
   'av_rech_amt_data_8', 'total_rech_data_8',
   'av_rech_amt_data_9', 'total_rech_data_9',
   'total_rech_amt_6', 'total_rech_amt_7',
    'total_rech_amt_8', 'total_rech_amt_9'
   ]].fillna(0)
```

##### Calculating total amount spent on data recharges for each month

```python
df['total_spend_data_6'] = df['av_rech_amt_data_6']*df['total_rech_data_6']
df['total_spend_data_7'] = df['av_rech_amt_data_7']*df['total_rech_data_7']
df['total_spend_data_8'] = df['av_rech_amt_data_8']*df['total_rech_data_8']
df['total_spend_data_9'] = df['av_rech_amt_data_9']*df['total_rech_data_9']
```

##### Calculating total amount spent (data + calls) for each month

```python
df['total_spend_calls_plus_data_6'] = df['total_spend_data_6']+df['total_rech_amt_6']
df['total_spend_calls_plus_data_7'] = df['total_spend_data_7']+df['total_rech_amt_7']
df['total_spend_calls_plus_data_8'] = df['total_spend_data_8']+df['total_rech_amt_8']
df['total_spend_calls_plus_data_9'] = df['total_spend_data_9']+df['total_rech_data_9']
```

```python
df['total_spend_calls_plus_data_6']
```

##### As per business understanding, 
- we know the months 6 and 7 are good phase months. 
- Further, we have to take average of these two months and filter data to keep the top 70% consumers only.

```python
df['total_spend_amt_good_phase'] = (df['total_spend_calls_plus_data_6']+df['total_spend_calls_plus_data_7'])/2
```

```python
df[['total_spend_calls_plus_data_6', 'total_spend_calls_plus_data_7', 'total_spend_amt_good_phase']].head()
```

##### selecting high value customers based on business definition

```python
df['total_spend_amt_good_phase'].quantile(0.7)
```

```python
df[df['total_spend_amt_good_phase'] > df['total_spend_amt_good_phase'].quantile(0.7)].shape
```

```python
df = df[df['total_spend_amt_good_phase'] > df['total_spend_amt_good_phase'].quantile(0.7)]
```

### Labelling Churn
- As per business understanding, if a customer has not used any data or calling services in month of september, then we will identify him as churn (1) else not churn (0)

```python
df['total_usage_9'] = df['total_ic_mou_9']+df['total_og_mou_9']+df['vol_2g_mb_9']+df['vol_3g_mb_9']
```

```python
df['churn'] = (df['total_usage_9']>0).map({True: 1, False: 0})
```

##### Calculating number of days since last recharge

```python
df['num_days_last_recharge_calls_6'] = df['last_date_of_month_6'] - df['date_of_last_rech_6']
df['num_days_last_recharge_calls_7'] = df['last_date_of_month_7'] - df['date_of_last_rech_7']
df['num_days_last_recharge_calls_8'] = df['last_date_of_month_8'] - df['date_of_last_rech_8']

df['num_days_last_recharge_data_6'] = df['last_date_of_month_6'] - df['date_of_last_rech_data_6']
df['num_days_last_recharge_data_7'] = df['last_date_of_month_7'] - df['date_of_last_rech_data_7']
df['num_days_last_recharge_data_8'] = df['last_date_of_month_8'] - df['date_of_last_rech_data_8']
```
